## Test for stationarity of the udiff series using the Dickey-Fuller test
With a p-value < 0.05, we can reject the null hypotehsis. This data set is stationary.

Looks like the weekly close is not stationary. (*without taking difference*)


```python
# Perform Dickey-Fuller test
dftest = sm.tsa.adfuller(udiff.weekly_ret, autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
for key, value in dftest[4].items():
    dfoutput['Critical Value ({0})'.format(key)] = value

dfoutput
```

```python
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

import pandas_datareader.data as web
import numpy as np

import statsmodels.tsa.api as tsa
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import acf, q_stat, adfuller
from scipy.stats import probplot, moment

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
def plot_correlogram(x, lags=None, title=None):
    lags = min(10, int(len(x)/5)) if lags is None else lags
    with sns.axes_style('whitegrid'):
        fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))
        x.plot(ax=axes[0][0], title='Residuals')
        x.rolling(21).mean().plot(ax=axes[0][0], c='k', lw=1)
        q_p = np.max(q_stat(acf(x, nlags=lags), len(x))[1])
        stats = f'Q-Stat: {np.max(q_p):>8.2f}\nADF: {adfuller(x)[1]:>11.2f}'
        axes[0][0].text(x=.02, y=.85, s=stats, transform=axes[0][0].transAxes)
        probplot(x, plot=axes[0][1])
        mean, var, skew, kurtosis = moment(x, moment=[1, 2, 3, 4])
        s = f'Mean: {mean:>12.2f}\nSD: {np.sqrt(var):>16.2f}\nSkew: {skew:12.2f}\nKurtosis:{kurtosis:9.2f}'
        axes[0][1].text(x=.02, y=.75, s=s, transform=axes[0][1].transAxes)
        plot_acf(x=x, lags=lags, zero=False, ax=axes[1][0])
        plot_pacf(x, lags=lags, zero=False, ax=axes[1][1])
        axes[1][0].set_xlabel('Lag')
        axes[1][1].set_xlabel('Lag')
        fig.suptitle(title, fontsize=14)
        sns.despine()
        fig.tight_layout()
        fig.subplots_adjust(top=.9)
```

### Find optimal $(p, q)$ lag

```python
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

from tqdm import tqdm
from itertools import product
import pandas as pd
import pandas_datareader.data as web
import numpy as np
from numpy.linalg import LinAlgError



from statsmodels.tsa.arima.model import ARIMA

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import acf, q_stat, adfuller
from scipy.stats import probplot, moment
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_size = 30
results = {}
y_true = udiff.iloc[train_size:].values
for p in range(3):
    for q in range(10):
        aic, bic = [], []
        if p == 0 and q == 0:
            continue
        print(p, q)
        convergence_error = stationarity_error = 0
        y_pred = []
        for T in range(train_size, len(udiff)):
            train_set = udiff.iloc[T-train_size:T]
            print(train_set)
            try:
                #print("fitting")
                model = ARIMA(train_set, order=(p, 0, q)).fit()
                #print(model)
                #print("fitting completed ", p, " ", q, " ", model)
            except LinAlgError:
                convergence_error += 1
            except ValueError:
                stationarity_error += 1

            forecast = model.forecast(steps=1)
            print("[INFO] prediction ", forecast.iloc[0], " ", p, " ", q )
            y_pred.append(forecast.iloc[0])
            aic.append(model.aic)
            bic.append(model.bic)
        print(y_true[0])
        print(y_true[1])
        y_true_flat = []
        for item in y_true:
          y_true_flat.append(item)
        y_true = y_true_flat
        print("y_true ", y_true)
        print("y_pred ", y_pred)
        result = (pd.DataFrame({'y_true': y_true, 'y_pred': y_pred})
                  .replace(np.inf, np.nan)
                  .dropna())

        rmse = np.sqrt(mean_squared_error(
            y_true=result.y_true, y_pred=result.y_pred))

        results[(p, q)] = [rmse,
                           np.mean(aic),
                           np.mean(bic),
                           convergence_error,
                           stationarity_error]
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error  # Import mean_squared_error

train_size = len(udiff) - 10 # 10 weeks of training data
results = {}
test_set = udiff.iloc[train_size:]
print(test_set)
from numpy.linalg import LinAlgError  # Import LinAlgError

for p1 in range(17, 21):
    for q1 in range(0, 1):
        for p2 in range(0, 1):
            for q2 in range(0, 1):
              preds = test_set.copy().assign(y_pred=np.nan)
              print(preds)
              preds['y_true'] = preds['weekly_ret']

              print(preds)
              aic, bic = [], []
              if p1 == 0 and q1 == 0:
                  continue
              convergence_error = stationarity_error = 0
              y_pred = []
              for i, T in enumerate(range(train_size, len(udiff))):
                  train_set = udiff.iloc[T-train_size:T]
                  try:
                      with warnings.catch_warnings():
                          warnings.filterwarnings("ignore")
                          print("[INFO] {:d} {:d} {:d} {:d} | {:3d} | [{:4d}-{:4d}]".format(p1, q1, p2, q2, i, T - train_size, T))
                          model = tsa.SARIMAX(endog=train_set.values,
                                          order=(p1, 0, q1),
                                          ).fit(disp=0)
                                         # seasonal_order=(p2, 0, q2, 12)
                  except LinAlgError:
                      convergence_error += 1
                  except ValueError:
                      stationarity_error += 1
                  print("[INFO] model.forecast = ", model.forecast(steps=1)[0])
                  preds.iloc[i, 1] = model.forecast(steps=1)[0]
                  print("[INFO] preds \n", preds, "========\n\n")
                  aic.append(model.aic)
                  bic.append(model.bic)

              preds.dropna(inplace=True)
              mse = mean_squared_error(preds.y_true, preds.y_pred)
              results[(p1, q1, p2, q2)] = [np.sqrt(mse),
                                                preds.y_true.sub(preds.y_pred).pow(2).std(),
                                                np.mean(aic),
                                                np.std(aic),
                                                np.mean(bic),
                                                np.std(bic),
                                                convergence_error,
                                                stationarity_error]
              print("[INFO] results \n", results, "\n\n")
              sarimax_results = pd.DataFrame(results).T
              sarimax_results.columns = ['RMSE', 'RMSE_std', 'AIC', 'AIC_std', 'BIC', 'BIC_std', 'convergence', 'stationarity']
              sarimax_results['CV'] = sarimax_results.RMSE_std.div(sarimax_results.RMSE)
              sarimax_results.index.names = ['p1', 'q1', 'p2', 'q2']
              sarimax_results.info()
              print("[INFO] Current best sarimax results \n")
              print(sarimax_results.nsmallest(20, columns='RMSE'))
              print("[INFO]===============\n\n")

```
